```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print("大人です")
        else:
            print("大人ではないです")

# インスタンス化する
person = [Human("自分", 25), Human("弟", 22), Human("父", 67), Human("母", 60), Human("架空の家族", 4)]

family = 0

while family < 5:
    # メソッドにアクセスして実行する
    person[family].check_adult()
    family = family+1
```
