```python
class Human:
    def __init__(self):
        self.name = ""
        self.age = ""

    # メソッドを定義する
    def set_name_age(self, name, age):
        self.name = name
        self.age = age

    def printinfo(self):
        print(self.name, self.age)

# インスタンス化する
nagafuchi = Human()

# メソッドにアクセスして実行する
nagafuchi.set_name_age("永渕", "25")
nagafuchi.printinfo()
```
